```
### your solution is here
import numpy as np
from numpy import linalg as LA
import matplotlib.pyplot as plt

def Laplacian(x):
    h = x[1]-x[0] # assume uniformly spaced points
    n = len(x)
    M = -2*np.eye(n)
    for i in range(1,n):
        M[i,i-1] = M[i-1,i] = 1
    return M/h**2
  
n = 101
x = np.linspace(-10,10,n)
T = -Laplacian(x)
V = x**2
H =  T + np.diag(V)
E,U = LA.eig(H)
h = x[1]-x[0]

# Plot the Harmonic potential
#plt.plot(x,V,color='k')

for i in range(5):
    # For each of the first few solutions, plot the energy level:
    plt.plot(x, [E[i]]*n,'k--')
    # axhline(y=E[i],color='k',ls=":")
    # as well as the eigenfunction, displaced by the energy level so they don't
    # all pile up on each other:
    plt.plot(x,U[:,i]/np.sqrt(h)+E[i])
plt.title("Eigenfunctions of the Quantum Harmonic Oscillator", fontsize=15)
plt.xlabel("Grid", fontsize=15)
plt.ylabel(r"$\frac{2E}{h\omega}$", fontsize=22)
plt.show()
###
```

```
xx = np.arange(10)
plt.plot(xx, 2*xx+1, "o--", label="Theoretical")
plt.plot(xx, E[:10], "o--", label="Numerical")
plt.title("Theoretical and numerical first 10 energy levels", fontsize=15)
plt.ylabel(r"$\frac{2E}{h\omega}$", fontsize=22)
plt.xlabel("Level", fontsize=15)
plt.grid(True)
plt.legend(fontsize=15)
plt.yticks(2*xx+1)
plt.show()
```
