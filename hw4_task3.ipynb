```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline

X = np.array([[1,1,2,3,5],[5,3,4,10,7],[2,1,2,5,5]])
```

```python
fig = plt.figure(figsize=(8, 6))
ax = Axes3D(fig, elev=-150, azim=110)
ax.scatter(X[0], X[1], X[2], edgecolor='k', c="r");
```

```python
shift = X.mean(axis=1).reshape(-1,1)
u, s, vh = np.linalg.svd(X-shift)

s2 = s**2
s2 /= s2.sum()
print ("ratio of explained variance: ", s2[:2].sum())
```

```python
vectors = u[:, :2] + shift
print ("u0 = ", vectors[:,0])
print ("u1 = ", vectors[:,1])
```

Let's plot points and the best plane span on u1 and u2

```python
tt = np.linspace(0, 2*np.pi, 51)
xx = s[0]*np.cos(tt)
yy = s[1]*np.sin(tt)
```

```python
v = vectors[:, :1] * xx + vectors[:, 1:] * yy
v.shape
```

```python
fig = plt.figure(figsize=(8, 6))
ax = Axes3D(fig, elev=-150, azim=110)
ax.scatter(X[0], X[1], X[2], edgecolor='k', c="r", )
ax.plot(v[0], v[1], v[2], c="b")
plt.show()
```
